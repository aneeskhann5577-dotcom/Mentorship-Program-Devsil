        4.4 Hypothesis Testing

```python
from scipy.stats import ttest_ind
```

```python
left = df[df['Attrition']=='Yes']['MonthlyIncome']
stay = df[df['Attrition']=='No']['MonthlyIncome']
ttest_ind(left, stay)
```

### Conclusion:
The independent ttest shows that p-value (0.000005) is less than 0.05. Therefore we reject the null hypothesis. This indicates that there is a significant difference in salary between employees who left and those who stayed. Employees who left the company had lower average salaries compared to those who stayed.
